```python
#Iteration
#While Loops: Theory & Examples

#while <condition>:
  # <body>

#- Test condition; if True, run body then re-test.
#- Can run **zero or many** times
```

```python
# lost forest game
where = input("You're in the Lost Forest. Go left or right? ")
while where == "right":
  where = input("You're in the Lost Forest. Go left or right? ")
print("You got out of the Lost Forest")
```

```python
#Counter Pattern
# Use a **counter variable** to limit loop iterations
# Contrast with infinite loop ('while True') and how to break out of loop.
n = 0
while n < 5:
  print(n)
  n += 1
```

```python
n = 0
while n < 5:
  print(n)
  n = n + 1
```

```python
#FOR LOOPS: Theory & Examples
#Python's range() function returns a sequence of numbers.

#for variable in <iterable>:
  #<body>
for i in range(5):      #0,1,2,3,4
  print(i)
for i in range(1,6):    #1,2,3,4,5
    print(i)
for i in range(1,10,2): #1,3,5,7,9
    print(i)
```

```python
fact = 1
x = int(input("Enter a number: "))
for i in range(1, x+1):
  fact = fact * i
print(f"{x}! = {fact}")
```

```python
#Given N,print "Hello World" on separate lines using for loop.
for i in range(int(input())):
    print("Hello World")
```

```python
while True:
  print("Hello World")
```

```python
#Early Exit With break
mysum = 0
for i in range(7, 101, 7):
  mysum += i
  if mysum > 100:
    break
print(mysum)
```

```python
mysum = 0
for i in range(5, 11):
  if mysum > 20:
    break
  mysum += i
print("Sum:" ,mysum)
```
